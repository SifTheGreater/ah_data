### Decks data

```python
from datetime import date, timedelta

start = '2016-09-01'
end = '2022-01-22'

start_date = date.fromisoformat(start)
end_date = date.fromisoformat(end)

date_range = [
    # end_date - timedelta(days=i)  # For date objects
    (end_date - timedelta(days=i)).isoformat()  # For ISO-8601 strings
    for i
    in range((end_date - start_date).days)
]
dates = list(reversed(date_range))
dates
```

```python
##getting all of the deck ids.
good_dates = list()
deck_ids = list()
for date in dates:
    print(date)
    data = rq.get(decks_base_url.format(date = date))
    if data.status_code == 200:
        good_dates.append(date)
        for x in data.json():
            deck_ids.append(x["id"])
```

```python
## saving the dates as a text file
with open("dates.txt", 'w') as file:
        for row in good_dates:
            s = "".join(map(str, row))
            file.write(s+'\n')
            
## saving the ids as a text file
with open("deck_ids.txt", "w") as output:
    output.write(str(deck_ids)[1:-1])
```

```python
## loading the dates from the text file
with open("dates.txt", "r") as input:
    text = input.read()
data_into_list = text.split("\n")
dates = data_into_list[:-1]

## loading the ids from the text file
with open("deck_ids.txt", "r") as input:
    text = input.read()
data_into_list = text.split(",")
ids = list(map(int, data_into_list))
```

```python
print(len(good_dates), len(deck_ids))
print(len(dates), len(ids))
print(good_dates == dates)
print(deck_ids == ids)
```
